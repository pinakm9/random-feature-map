```python
import numpy as np 
from scipy.integrate import odeint
import os, sys, warnings
from pathlib import Path
from os.path import dirname, realpath
script_dir = Path(dirname(realpath('.')))
module_dir = str(script_dir)
sys.path.insert(0, module_dir + '/modules')
import utility as ut
import surrogate1 as sr1
import surrogate as sr
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
L0, L1, n, s, lw = 0.4, 3.5, 100, 0.1, 0.5
o = np.ones(n)
c = ['#1f77b4', '#2ca02c', '#ff7f0e']
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)
x_g = np.concatenate([np.linspace(-L1, -L0, num=n), np.linspace(L0, L1, num=n)], axis=0)
x_l = np.linspace(-L0, L0, num=n)
x_e = np.concatenate([np.linspace(-L1-3, -L1, num=n), np.linspace(L1, L1+3, num=n)], axis=0)
ax.plot(x_g[:n], np.tanh(x_g[:n]), c=c[0], label='good')
ax.plot(x_g[n:], np.tanh(x_g[n:]), c=c[0])
ax.plot(x_l, np.tanh(x_l), c=c[2], label='linear', linestyle='dotted')
ax.plot(x_e[:n], np.tanh(x_e[:n]), c=c[1], linestyle='dashed')
ax.plot(x_e[n:], np.tanh(x_e[n:]), c=c[1], label='saturated', linestyle='dashed')
# ax.plot(-L0*o, np.linspace(-1, np.tanh(-L0), n), c='black')
ax.axvline(-L0, c='black', linewidth=lw)
ax.axvline(L0, c='black', linewidth=lw)
ax.axvline(-L1, c='black', linewidth=lw)
ax.axvline(L1, c='black', linewidth=lw)
ax.set_ylabel(r'$\tanh(z)$', fontsize=13)
ax.set_xlabel(r'z', fontsize=13)
# ax.set_title('tanh')
ax.set_xticks([-L1-3, -L1, -L0, L0, L1, L1+3])
xticks = [item.get_text() for item in ax.get_xticklabels()]
xticks[1] = '$-L_1$' 
xticks[2] = '$-L_0$'
xticks[3] = '$L_0$' 
xticks[4] = '$L_1$' 
ax.set_xticklabels(xticks)
ax.legend(fontsize=13)
ax.set_xticklabels(xticks, fontsize=15)
plt.tight_layout()
plt.savefig('../data/plots/tanh.png', dpi=300)
plt.show()
```
