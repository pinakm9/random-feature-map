```python
data = bsa.get_data()
zc_g = np.zeros_like(arr_a)
zc_vg = np.zeros_like(arr_a)
num_vg = np.zeros_like(arr_a)
num_g = np.zeros_like(arr_a)
zc_b = np.zeros_like(arr_a)
num_b = np.zeros_like(arr_a)
for i, a in enumerate(arr_a):
    arr = data.loc[(data['n_good_rows_W_in'] == int(a)) & (data['tau_f_se'] > 4.)]['0-cols-W'] / bsa.D_r
    zc_g[i] = arr.mean()
    num_g[i] = len(arr)
    arr_ = data.loc[(data['n_good_rows_W_in'] == int(a)) & (data['tau_f_se'] < 1.5)]['0-cols-W'] / bsa.D_r
    zc_b[i] = arr_.mean()
    num_b[i] = len(arr_)
    arr__ = data.loc[(data['n_good_rows_W_in'] == int(a)) & (data['tau_f_se'] > 5.)]['0-cols-W'] / bsa.D_r
    zc_vg[i] = arr__.mean()
    num_vg[i] = len(arr__)


plt.plot(arr_a/bsa.D_r, zc_g, label='generalize')
plt.plot(arr_a/bsa.D_r, zc_vg, label='superb', linewidth=1)
plt.plot(arr_a/bsa.D_r, zc_b, label='overfit')

plt.xlabel('good rows')
plt.ylabel('average zero cols W')
plt.legend()
```

```python
plt.plot(arr_a/bsa.D_r, num_g, label='generalize')
plt.plot(arr_a/bsa.D_r, num_b, label='overfit')
plt.xlabel('good rows')
plt.ylabel('number of occurrences')
plt.legend()
```

```python
from scipy import stats

save_folder = '../data/L63-Strategy-SMLR-1'
bsa = sr.BatchStrategyAnalyzer_SMLR(save_folder)
dx = 100./50
bsa.percents = np.arange(0., 100. + dx, dx)
# bsa.percents = np.array()
n_rows, n_cols = 5, 2
fig = plt.figure(figsize=(4*n_cols, 4*n_rows))
axes = [fig.add_subplot(n_rows, n_cols, j+1) for j in range(5)]
quantities = ['0-cols-W', '||W||']
labels = ['zero cols W', '||W||']

for i in range(2, 4, 2):
    j = int(i/2)
    mean, std, num = bsa.get_generalize_overfit_superb(quantities[j])
    if i > 0:
        mean = [e/bsa.D_r for e in mean]
        std = [e/bsa.D_r for e in std]

    axes[i].plot(np.log(bsa.percents/100.), np.log(mean[0]), label='generalize') 
    axes[i].plot(np.log(bsa.percents/100.), np.log(mean[1]), label='overfit') 
    axes[i].plot(np.log(bsa.percents/100.), np.log(mean[2]), label='superb')

    axes[i].scatter(np.log(bsa.percents/100.), np.log(mean[0]), label='generalize') 
    axes[i].scatter(np.log(bsa.percents/100.), np.log(mean[1]), label='overfit') 
    axes[i].scatter(np.log(bsa.percents/100.), np.log(mean[2]), label='superb')

    X = np.log(bsa.percents[30:]/100.)
    y = np.log(mean[1][30:]/1)
    res = stats.linregress(X, y)
    print(labels[j], res.slope, res.intercept)

    X = np.log(bsa.percents[30:]/100.)
    y = np.log(std[1][30:]/1)
    res = stats.linregress(X, y)
    print(labels[j] + 'std', res.slope, res.intercept)

    axes[i].set_xlabel('good rows')
    axes[i].set_ylabel(labels[j])
    axes[i].legend()

    axes[i+1].plot(np.log(bsa.percents/100.), np.log(std[0]), label='generalize') 
    axes[i+1].plot(np.log(bsa.percents/100.), np.log(std[1]), label='overfit') 
    axes[i+1].plot(np.log(bsa.percents/100.), np.log(std[2]), label='superb') 

    axes[i+1].scatter(np.log(bsa.percents/100.), np.log(std[0]), label='generalize') 
    axes[i+1].scatter(np.log(bsa.percents/100.), np.log(std[1]), label='overfit') 
    axes[i+1].scatter(np.log(bsa.percents/100.), np.log(std[2]), label='superb') 

    axes[i+1].set_xlabel('good rows')
    axes[i+1].set_ylabel('std ' + labels[j])
    axes[i+1].legend()

axes[-1].plot(np.log(bsa.percents/100.), np.log(num[0]/1), label='generalize') 
axes[-1].plot(np.log(bsa.percents/100.), np.log(num[1]/1), label='overfit') 
axes[-1].plot(np.log(bsa.percents/100.), np.log(num[2]/1), label='superb') 

axes[-1].scatter(np.log(bsa.percents/100.), np.log(num[0]/1), label='generalize') 
axes[-1].scatter(np.log(bsa.percents/100.), np.log(num[1]/1), label='overfit') 
axes[-1].scatter(np.log(bsa.percents/100.), np.log(num[2]/1), label='superb') 


X = np.log(bsa.percents[30:]/100.)
y = np.log(num[1][30:]/1)
res = stats.linregress(X, y)
print(res.slope, res.intercept)








axes[-1].set_xlabel('good rows')
axes[-1].set_ylabel('number of occurences')
axes[-1].legend()

fig.subplots_adjust(wspace=0.3, hspace=0.5)
plt.show()
```
